```python
df = pd.read_csv("Base de datos economía circular.csv")
```

```python
df = df[["parentesco_jefe_hogar", "estado_civil","viaje_noche_fuera","nivel_educativo","edad","estrato","num_viajes_12_meses"]]
```

```python
df
```

```python
df = df[df["parentesco_jefe_hogar"] == "jefe_hogar"]
```

```python
df = df[~df["viaje_noche_fuera"].isna()]
```

```python
df = df[~df["num_viajes_12_meses"].isna()]
```

```python
df = df[df["estrato"] < 9.0]
```

```python
df["viaje_noche_fuera"]=df["viaje_noche_fuera"].replace({"si": 1, "no": 0}) 
```

```python
df["num_viajes_12_meses"]=df["num_viajes_12_meses"].replace({"si": 1, "no": 0})
```

```python
df.drop(columns="parentesco_jefe_hogar",inplace=True)
```

```python
df = df[df["nivel_educativo"]!= "no_sabe_no_informa"]
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
X = df[["edad", "estrato", "estado_civil_pareja_union_libre", "nivel_educativo_superior_universitaria", "viaje_noche_fuera"]].copy()
y = X.pop("viaje_noche_fuera")
X = sm.add_constant(X)
model = sm.OLS(y,X)
reg = model.fit()
reg.summary()
```

```python
X = df[["edad", "estrato", "estado_civil_pareja_union_libre", "nivel_educativo_superior_universitaria", "num_viajes_12_meses"]].copy()
y = X.pop("num_viajes_12_meses")
X = sm.add_constant(X)
model = sm.OLS(y,X)
reg = model.fit()
reg.summary()
```
